```python
def process(axs, rowIndex, colIndex, templateName, filename: str=None) -> None:
    """
    View multiple images stored in files, stacking vertically

    Arguments:
        filename: str - path to filename containing image
    """
    image = mpimg.imread(filename)
    # <something gets done here>
    axs[rowIndex, colIndex].set_title(templateName.lower().capitalize(), fontweight='bold')
    axs[rowIndex, colIndex].imshow(image, interpolation='none')
    axs[rowIndex, colIndex].axis('off')
```

```python
import imageio
import numpy as np    

#Create reader object for the gif
gif1 = imageio.get_reader('2500-rl-benchmark/assets/images/rendered/gifs/roof.gif')
gif2 = imageio.get_reader('2500-rl-benchmark/assets/images/rendered/gifs/firctionbalanceball.gif')

#If they don't have the same number of frame take the shorter
number_of_frames = min(gif1.get_length(), gif2.get_length()) 

#Create writer object
new_gif = imageio.get_writer('2500-rl-benchmark/assets/images/rendered/gifs/combined.gif')

for frame_number in range(number_of_frames):
    img1 = gif1.get_next_data()
    img2 = gif2.get_next_data()
    #here is the magic
    new_image = np.hstack((img1, img2))
    new_gif.append_data(new_image)

gif1.close()
gif2.close()    
new_gif.close()
```
